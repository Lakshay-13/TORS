```python
def eval(s):

    s = s.strip('[]').strip()
    string_list = s.split(',')
    float_list = [float(x) if x != 'NaN' else 999 for x in string_list]
    
    return float_list
```

```python
class TORS:
    
    def __init__(self, path):
        
        self.path = path
        
    def serialize_gaia(self, dictionary):

        dictionary_features={
        'id': tf.train.Feature(bytes_list=tf.train.BytesList(value=[str(dictionary['source_id']).encode()])),
        'ra': tf.train.Feature(bytes_list=tf.train.BytesList(value=[str(dictionary['ra']).encode()])),
        'dec': tf.train.Feature(bytes_list=tf.train.BytesList(value=[str(dictionary['dec']).encode()])),
        'class': tf.train.Feature(bytes_list=tf.train.BytesList(value=[str(dictionary['class']).encode()])),
        'transit_id': tf.train.Feature(float_list=tf.train.FloatList(value=dictionary['transit_id'])),
        'g_obs_time': tf.train.Feature(float_list=tf.train.FloatList(value=dictionary['g_obs_time'])),   
        'g_mag': tf.train.Feature(float_list=tf.train.FloatList(value=dictionary['g_mag'])),
        'bp_obs_time': tf.train.Feature(float_list=tf.train.FloatList(value=dictionary['bp_obs_time'])),
        'bp_mag': tf.train.Feature(float_list=tf.train.FloatList(value=dictionary['bp_mag'])),
        'rp_obs_time': tf.train.Feature(float_list=tf.train.FloatList(value=dictionary['rp_obs_time'])),
        'rp_mag': tf.train.Feature(float_list=tf.train.FloatList(value=dictionary['rp_mag']))
        }


        element_context = tf.train.Features(feature = dictionary_features)

        example = tf.train.SequenceExample(context= element_context)

        return example
    
    def get_gaia_dataset(self, links_file, record_file):
        
        links_path = self.path + '\\' + links_file
        with open(links_path, 'r') as f:
            links = f.read().split()
               
        necessary_columns = ['solution_id', 'source_id', 'transit_id', 'g_transit_time', 'g_transit_mag', 'bp_obs_time', 'bp_mag', 'rp_obs_time', 'rp_mag']
        vari_stars = pl.read_parquet(self.path + '\\variable_stars_with_pos.parquet')
        unread_files = []
        
        with open(f'{record_file}.tfrecord', 'w') as f:
            writer = tf.io.TFRecordWriter(f.name)
            for i in tqdm_notebook(links):
                try:
                    pdf = pl.read_csv(i, comment_char='#')
                    pdf = pdf[necessary_columns]
                    pdf = pdf.join(vari_stars, on = 'source_id')
                    pdf = pdf.drop(['solution_id', 'solution_id_right'])
                    for j in range(pdf.shape[0]):
                        temp = {}
                        temp['source_id'] = str(pdf['source_id'][j]) #i64
                        temp['ra'] = str(pdf['ra'][j]) #f64
                        temp['dec'] = str(pdf['dec'][j]) #f64
                        temp['class'] = pdf['class'][j] #string
                        temp['transit_id'] = eval(pdf['transit_id'][j]) #i64 list
                        temp['g_obs_time'] = eval(pdf['g_transit_time'][j]) #f list
                        temp['g_mag'] = eval(pdf['g_transit_mag'][j]) #f list
                        temp['bp_obs_time'] = eval(pdf['bp_obs_time'][j]) #f list
                        temp['bp_mag'] = eval(pdf['bp_mag'][j]) #f list
                        temp['rp_obs_time'] = eval(pdf['rp_obs_time'][j]) #f list
                        temp['rp_mag'] = eval(pdf['rp_mag'][j]) #f list
                        ex = self.serialize_gaia(temp)
                        writer.write(ex.SerializeToString())
                except:
                    unread_files.append(i)
                    print(f'File {i} not read')

            with open('unread_files.dat', 'w') as f:
                for i in unread_files:
                    f.write(i + '\n')
        print(f'Dataset saved as {record_file}.tfrecord')
```

```python
class TORS_RL:
    
    """
    A class representing the TORS_RL reinforcement learning agent.
    
    Attributes:
        action_space (int): Number of possible actions the agent can take.
        state (dict): Current state of the agent, including position, and G, RP, and BP band data.
        single_band_model (model): Trained single-band model for making predictions.
        multiband_model (model): Trained multiband model for making predictions.
        reward (float): Cumulative reward of the agent.
    """
    
    def __init__(self, single_band_model, multiband_model):
        
        """
        Initializes the TORS_RL agent with the given single-band and multiband models.
        
        Args:
            single_band_model (model): Trained single-band model for making predictions.
            multiband_model (model): Trained multiband model for making predictions.
        """
        
        self.action_space = 3  # 0: No observation, 1: G band, 2: Multiband
        self.state = {
            'position': (None, None),  # (ra, dec)
            'XG': np.zeros(3),
            'XRP': np.zeros(3),
            'XBP': np.zeros(3)
        }
        self.single_band_model = single_band_model
        self.multiband_model = multiband_model
        self.reward = 0

    def step(self, action):
        
        """
        Updates the state and reward based on the given action.
        
        Args:
            action (int): Action taken by the agent (0: No observation, 1: G band, 2: Multiband).
        
        Returns:
            tuple: Updated state and reward.
        """
        
        assert 0 <= action < self.action_space, "Invalid action."

        prev_probability = self.get_probability(action)

        next_state = self.update_state(action)
        next_probability = self.get_probability(action)

        if action == 0:
            self.reward += 0
        elif action == 1:
            self.reward += 10 * (next_probability - prev_probability)
        elif action == 2:
            self.reward += 3 * (next_probability - prev_probability)

        self.state = next_state
        return self.state, self.reward

    def get_probability(self, action):
        
        """
        Calculates the probability for the given action using the appropriate model.
        
        Args:
            action (int): Action taken by the agent (0: No observation, 1: G band, 2: Multiband).
        
        Returns:
            float: Probability for the given action.
        """
        
        if action == 1:
            return self.single_band_model.predict(self.state['XG'].reshape(1, 3, 1))[0][0]
        elif action == 2:
            return self.multiband_model.predict([self.state['XG'].reshape(1, 3, 1), self.state['XRP'].reshape(1, 3, 1), self.state['XBP'].reshape(1, 3, 1)])[0][0]
        else:
            return 0
        
    def update_state(self, action):
        
        """
        Updates the agent's state based on the given action.
        
        Args:
            action (int): Action taken by the agent (0: No observation, 1: G band, 2: Multiband).
        
        Returns:
            dict: Updated state.
        """
        
        next_state = self.state.copy()

        # Update the state based on the action taken
        if action == 1:
            next_state['XG'] = self.observe_g_band()
        elif action == 2:
            next_state['XG'], next_state['XRP'], next_state['XBP'] = self.observe_multiband()

        return next_state

    def observe_g_band(self):
        
        """
        Observes the G band and returns the updated G band data.
        
        Returns:
            np.array: Updated G band data.
        """
        
        return np.zeros(3)

    def observe_multiband(self):
        
        """
        Observes all three bands (G, RP, and BP) and returns the updated band data.
        
        Returns:
            tuple: Updated G, RP, and BP band data.
        """
        
        return (np.zeros(3),np.zeros(3),np.zeros(3))

    def set_position(self, ra, dec):
        
        """
        Sets the position of the agent in the state.
        
        Args:
            ra (float): Right ascension.
            dec (float): Declination.
        """
        
        self.state['position'] = (ra, dec)

    def reset(self):
        
        """
        Resets the state and reward of the agent.
        """
        
        self.state = {
            'position': (None, None),
            'XG': np.zeros(3),
            'XRP': np.zeros(3),
            'XBP': np.zeros(3)
        }
        self.reward = 0 
```

```python
def create_single_band_lstm_model():
    input_layer = Input(shape=(3, 1))
    lstm_layer = LSTM(32)(input_layer)
    output_layer = Dense(1, activation='sigmoid')(lstm_layer)
    model = Model(inputs=input_layer, outputs=output_layer)
    return model

def create_multiband_lstm_model():
    input1 = Input(shape=(3, 1))
    input2 = Input(shape=(3, 1))
    input3 = Input(shape=(3, 1))

    concat_inputs = Concatenate(axis=-1)([input1, input2, input3])

    lstm_layer = LSTM(32)(concat_inputs)
    output_layer = Dense(1, activation='sigmoid')(lstm_layer)

    model = Model(inputs=[input1, input2, input3], outputs=output_layer)
    return model

def random_update(arr):
    return arr + np.random.normal(size=arr.shape)

env = TORS_RL(create_single_band_lstm_model(), create_multiband_lstm_model())

env.set_position(ra=random.uniform(0, 360), dec=random.uniform(-90, 90))

num_steps = 1

for _ in range(num_steps):
    action = random.randint(0, 2) 
    next_state, reward = env.step(action)

    if action == 1:
        next_state['XG'] = random_update(next_state['XG'])
    elif action == 2:
        next_state['XG'] = random_update(next_state['XG'])
        next_state['XRP'] = random_update(next_state['XRP'])
        next_state['XBP'] = random_update(next_state['XBP'])

    print(f"Action: {action}, Next state: {next_state}, Reward: {reward}")

env.reset()
```
